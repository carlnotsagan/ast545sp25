Trends in the exoplanet population as function of stellar mass and metallicity, illustrating the different behavior of the giant planet population (**large blue circles**) and planets smaller than Neptune (**small red circles**). Credit: [Mulders et al 2024](https://arxiv.org/html/1805.00023v2#Ch0.S1).

**Key regimes**

1. Low stellar mass and low metallicity: primarily sub-Neptune planets. 
2. Low stellar mass and high metallicity: primarily sub-Neptune planets and _some_ giant planets. 
3. High stellar mass and low metallicity: Some sub-Neptune planets and _some_ giant planets.
4. High stellar mass and high metallicity: Mixture of sub-Neptune planets and giant planets.


### Positive Giant Planet-Metallicity Correlation


```{image} ./np_vs_Z.png
:align: center
:width: 80%
:name: np_vs_Z
```

Giant planet occurrence rate as function of stellar metallicity. Credit: Fischer and Valenti (2005).

````{prf:observation}
Giant planets occur more frequently around stars with higher metallicities.
````

One of the leading theories to describe this trend is that of primordial origin:

1. The metallicity of the host star is representative of the primordial cloud from which it formed. 
2. Implying that in a high-metallicity protoplanetary disk, giant planets form more easily. 

This theory is supported by the **core accretion** planet formation theory:

Recall:

**core accretion** - a bottom-up process:

1. the formation of giant planets begins with a rocky/icy core (10–15 $M_{\oplus}$)
2. gas is accreted onto the core in a runaway process, until it has cleared its orbit or the gas is removed from the disk. 
3. if this disk has a higher metallicity (more grains), the metal cores would be more easily built before the gas in the disk is lost. 


>The _Giant planet–metallicity correlation_ is thus an important piece of observational evidence in support of this scenario. [Osborn & Bayliss, 2019](https://academic.oup.com/mnras/article/491/3/4481/5628339). Less massive planets (sub-Neptune) show a wider range of distributions accross metallicity. 


## Trends With Stellar Mass

The correlation between planet occurrence and stellar mass is dependent on planet size.



```{image} ./np_vs_mass.png
:align: center
:width: 80%
:name: np_vs_mass
```
Giant planet occurrence as function of stellar mass. Credit: Johnson et al. (2010).

> Giant planets occur more frequently around higher-mass stars. Sub-Neptunes, those found in abundance with the Kepler survey, occur more frequently around low-mass M dwarfs.

We can look at this in more detail at a bifurcation mass of $M_{host}\sim0.34M_{\odot}$:


**$M_{host}\lt0.34M_{\odot}$**
```{image} ./low_mass.png
:align: center
:width: 70%
:name: low_mass_pop
```

**$M_{host}\gt0.34M_{\odot}$**
```{image} ./high_mass.png
:align: center
:width: 70%
:name: high_mass_pop
```

We can observe few giant planets for $M_{host}\gt0.34M_{\odot}$ in our sample, with sub-Neptune planets abundantly in systems with less massive M dwarf host stars $M_{host}\lt0.34M_{\odot}$. These planets also tend to be smaller and orbit closer to the host star. 

## In-Class Assignment 22

In-Class Assignment 22 can be found [here](../../assignments/ica22.ipynb).
